```python
# On saisit le jeu de donnees
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
# On fait un pretraitemnt avant d'appliquer apriori
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Appliquons le one-hot-encoding a notre jeu de donnee
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creation d'un nouveau DataFrame a partir de notre tableau Numpy
df
```

```python
# On applique la methode apriori avec la metric support
from mlxtend.frequent_patterns import apriori
frequent_itemsets1=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets1
```

```python
# On applique la methode apriori avec la metric confidence
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets1,metric="confidence",min_threshold=0.7) # association des itemsets avec une confidence d'au moins 70%.
```

```python
# On applique la methode apriori avec la metric lift
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets1,metric="lift",min_threshold=1.25)
```

```python
# Lecture du nouveau jeu de donnee
data=pd.read_csv('Market_Basket_Optimisation.csv')
data
```

```python
#Supression des colonnes qui n'ont aucune donnee
data=data.dropna(axis=1,how='all')
data
```

```python
#On remplace les donnees manquantes du dataframe par la mode
data.fillna(data.mode,axis=1,inplace=True)
```

```python
#Verifions 
data.isnull().sum()
```

```python
# On fait un pretraitemnt avant d'appliquer apriori
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)    #Appliquons one-hot-encoding a notre df
data=pd.DataFrame(te_ary, columns=te.columns_)  #Creation d'un nouveau DataFrame a partir du df
data
```

```python
# On applique la methode apriori avec la metric support
frequent_itemsets=apriori(data, min_support=0.6, use_colnames=True) #prise en charge des noms des colonnes
frequent_itemsets
```

```python
# On applique la methode apriori avec la metric confidence
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # association des itemsets une confidence d'au moins  70%.
```

```python
# On applique la methode apriori avec la metric lift
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
